# 绘制简单的函数曲线

所有图中最简单的是单个函数 y = f(x) 的可视化。这里，我们首先来看看如何创建这种类型的简单图。

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
#plt.style.use('seaborn-whitegrid') # 设置绘图风格
import numpy as np
seaborn.set()
```

* 1. 我们首先创建一个画布以及它的轴实例

```python
help(plt.plot)
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
```

* 2. **绘图**：两种方式：
* 1）面向对象式——所有的绘图方法都是基于轴类对象的。

```python
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig,ax = plt.subplots()
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend(frameon=False)  # Add a legend.
```

* 2) 直接使用plt.plot()方法: 更常用的方法，建议大家使用这种

```python
x = np.linspace(0, 2, 100)
plt.subplots(figsize=(6,4),dpi=150)
plt.plot(x, x, label='linear',color='b')  
plt.plot(x, x**2, label='quadratic',color='g') 
plt.plot(x, x**3, label='cubic',color='r')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(8,6),dpi=80) # 可以通过设置figsize 和 dpi 参数设置画布的大小和像素值
plt.plot(x, x, label='linear')  # 自动生成一个 figure 和 axis
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

## 设置线条的样式


* 1. 设置颜色，plot()的 color 参数

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

如果颜色没有设置，matplotlib会自动赋予多个线条不同的颜色；

* 2. 设置样式，plot()的 lifestyle 参数

```python
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

也可以简写起来

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

 ## 选取坐标轴的范围

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

一个有用的相关方法是 plt.axis() (注意这里*axis*与*e*、*axis*与*i*之间的潜在混淆)。
 plt.axis() 方法允许通过一个指定 \[xmin, xmax, ymin, ymax\] 的列表，通过一次调用来设置 x 和 y 限制。

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal') # 自动收紧边界
```
