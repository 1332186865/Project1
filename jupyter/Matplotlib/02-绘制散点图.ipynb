# 简单散点图的绘制

另一种常用的图类型是简单的散点图，是直线图的近亲。这里的点不是由线段连接，而是用点、圆或其他形状单独表示。

````python
%matplotlib inline
import matplotlib.pyplot as plt
#plt.style.use('seaborn-whitegrid')
import numpy as np
````

## 1. 使用 ``plt.plot``绘制散点图

在上一节中，我们看到了 plt.plot 来产生线，也可以产生散点图

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
````

用于标记点的形状的符号有很多

````python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=3)
````

这些符号可以与线和颜色符号一起使用：用一条线将点连接起来

````python
plt.plot(x, y, '-ok');
````

````python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
````

## 2. 使用 ``plt.scatter``绘制散点图

````python
plt.scatter(x, y, marker='o');
````

 ``plt.scatter`` 和 ``plt.plot`` 的主要区别在于 scatter 绘制的散点图中每个点的属性(大小、面颜色、边缘颜色等)可以单独控制或映射到数据。

````python
rng = np.random.RandomState(0) # 生成一个随机状态的种子，在相同种子下生成的随机数有相同的模式
x = rng.randn(100)
y = rng.randn(100)
print(x[:10],y[:10],sep='\n')
````

````python
colors = rng.rand(100)
colors[:10]
````

````python
sizes = 1000 * rng.rand(100)
sizes[:10]
````

````python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='rainbow') # cmap 是一种配色方案
plt.colorbar()  # show color scale
````

注意，color参数会自动映射到一个颜色比例(这里由 colorbar() 命令显示)，size参数以像素为单位给出。  
这样，点的颜色和大小就可以在可视化中传达信息，实现多维数据的可视化。

例如，我们可以使用来自Scikit-Learn的鸢尾花数据，其中每个样本都是仔细测量过花瓣和萼片大小的花。

````python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.data)
features = iris.data.T
````

````python
features = iris.data.T
plt.subplots(figsize=(6,4),dpi=80)
plt.scatter(features[0], features[1],marker = 'o',s=100*features[3], 
    c=iris.target,cmap = 'rainbow_r',alpha=0.3)
````

````python
plt.subplots(figsize=(6,4),dpi=80)
plt.scatter(features[0], features[1],marker = 'o',s=100*features[3],
            c=iris.target,cmap = 'rainbow_r', alpha=0.3)
plt.grid()
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
````

我们可以看到，这个散点图让我们能够同时探索数据的四个不同维度:  
每个点的(x, y)位置对应萼片的长度和宽度，点的大小与花瓣的宽度有关，颜色与花朵的特定种类有关。  
像这样的多颜色和多特征的散点图对于数据的探索和显示都很有用。
