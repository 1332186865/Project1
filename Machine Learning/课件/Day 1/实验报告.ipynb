## 实验1：随机漫步

### 实验目的： 通过模拟随机漫步学习如何运用数组运算
### 实验步骤：
+ 通过内置的random模块以纯Python的方式实现随机漫步并生成折线图
+ 使用np.random模块一次性随机产生
+ 一次模拟多个随机漫步

```python
import random

import pandas as pd

random.seed(5)
# 法1
position = 0
walk = [position]
steps = 100
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
print(walk)
# 法二
step2 = [1 if random.randint(0, 1) else -1 for _ in range(steps)]
walk2 = [sum(step2[:i + 1]) for i in range(steps)]
print(walk2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
sn.set()
# plt.subplot(2, 1, 2)
plt.plot(walk, label='walk')
plt.plot(walk2, label='walk2')
plt.legend()
plt.show()
```

```python
import numpy as np
# 法三
nsteps = 100
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0, 1, -1)
walk3 = steps.cumsum()
plt.plot(walk3, label='walk3')
plt.legend()
```

```python
# 一次生成多个
nwalks = 5
nsteps = 100
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(axis=1)
walks
```

```python
plt.plot(walks.T)
```

```python
steps = np.random.normal(loc=0, scale=0.25,size=(nwalks, nsteps))
normal_walk = steps.cumsum(axis=1)
normal_walk
```

```python
plt.plot(normal_walk.T, label='normal')
```

### 心得：
在数据处理上，numpy库确实比Python有更大的优势。

## 实验二：计算指标/哑变量
### 实验目的： 通过MovieLens 1M数据集体会如何对数据进行编码
### 实验步骤：
+ 从文件读取数据
+ 从数据中抽取特征
+ 通过迭代对数据进行编码


```python
import pandas as pd
import numpy as np
```

#### 1. 导入数据

```python
names = ['movie_id', 'title', 'genres']
movies = pd.read_table('movies.dat', sep='::',header=None, names=names, engine='python')
movies.head()
```

#### 2. 查看数据信息

```python
movies.describe()
```

```python
movies.info()
```

#### 3. 重设索引

```python
movies.set_index('movie_id', inplace=True)
movies
```

#### 4. 判断数据是否缺失

```python
movies.isnull().any()
```

#### 5. 判断数据是否重复

```python
movies.duplicated().any()
```

#### 6. 提取特征

```python
all_genres = []
for x in movies.genres:
    all_genres.extend(x.split('|'))
genres = pd.unique(all_genres)
genres
```

#### 7. 编码

```python
# 构建指标DataFrame
zero_matrix=np.zeros(shape=(len(movies), len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)
dummies.head()
```

```python
for num, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[num, indices] = 1
dummies.head()
```

```python
# 每种特征所包含的书籍数量
dummies.sum()
```

#### 8. 图像显示

```python
import seaborn as sn
import matplotlib.pyplot as plt
sn.set()
plt.figure(figsize=(60, 20), dpi=100)
plt.rcParams['font.family'] = 'SimHei'
plt.bar(dummies.sum().index, height=dummies.sum(), label=dummies.columns,width=0.5)
plt.xlabel('类型', fontsize=50)
plt.ylabel('数量', fontsize=50)
plt.xticks(fontsize=40)
plt.yticks(fontsize=30)
```
