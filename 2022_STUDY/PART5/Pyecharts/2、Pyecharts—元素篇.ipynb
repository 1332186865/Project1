### LegendOptsï¼šå›¾ä¾‹é…ç½®é¡¹

é’ˆå¯¹å›¾è¡¨å›¾ä¾‹çš„æ ·å¼è¿›è¡Œé…ç½®ï¼Œå¯é…ç½®å†…å®¹å¦‚ä¸‹ï¼š

### DataZoomOptsï¼šåŒºåŸŸç¼©æ”¾é…ç½®é¡¹

å½“åæ ‡è½´æ•°æ®é¡¹è¿‡å¤šæ—¶å€™ï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨ç¼©ç•¥è½´ï¼Œé…ç½®å†…å®¹å¦‚ä¸‹ï¼š

* ç¼©æ”¾ç±»å‹ï¼š**slider**å’Œ**inside**ä¸¤ç§ç±»å‹

* ç¼©æ”¾èŒƒå›´ï¼šå¯ä»¥æ ¹æ®å€¼ï¼Œä¹Ÿå¯ä»¥æ ¹æ®ç™¾åˆ†æ¯”è¿›è¡Œç¼©æ”¾ï¼›

* æ˜¯å¦é”ç‚¹ç¼©æ”¾åŒºåŸŸå¤§å°ï¼šå¦‚é”å®šåˆ™åªèƒ½è¿›è¡Œå¹³ç§»ï¼Œä¸èƒ½æ‰©å¤§é€‰æ‹©èŒƒå›´ï¼›

* å¸ƒå±€æ–¹å¼ï¼šå‚ç›´ or æ°´å¹³å¸ƒå±€ï¼Œè¿™ä¹Ÿå†³å®šäº†æ˜¯å¯¹äºxè½´è¿›è¡Œç¼©æ”¾è¿˜æ˜¯yè½´ç¼©æ”¾


````python
import random

# è™šå‡æ•°æ®
x_data = list(range(1990,2020))
y_data = [random.randint(0, 100) for _ in x_data]

chart = Bar()

chart.add_xaxis(x_data)

chart.add_yaxis(
    '',
    y_data
)

chart.reversal_axis()
chart.set_global_opts(
    datazoom_opts=opts.DataZoomOpts(
        range_start=50,  # å¼€å§‹èŒƒå›´
        range_end=80,  # ç»“æŸèŒƒå›´
        orient='vertical',  # è®¾ç½®ä¸ºå‚ç›´å¸ƒå±€
        type_='slider',  # sliderå½¢å¼
        is_zoom_lock=True,  # é”å®šåŒºåŸŸå¤§å°
        pos_left='1%'  # è®¾ç½®ä½ç½®
        )
)

chart.render_notebook()
````
